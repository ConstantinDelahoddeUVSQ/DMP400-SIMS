(Schéma)

On considère la plaque chargée uniformément en volume avec une charge Q.

A présent calculons E.

On considère $\vec{E}$ uniquement selon $\vec{e_y}$ et on néglige les effets de bord.
On se place donc dans l'approximation de la plaque infinie. 
Donc $\vec{E} = E \vec{e_y}$  

---

### Application du Théorème de Gauss :

$$
\oint_{S_1} \vec{E} \cdot \mathrm{d}\vec{S_1} \, \vec{e_y} 
+ \oint_{S_2} (-\vec{E}) \cdot (-\mathrm{d}\vec{S_2}) \, \vec{e_y} 
+ \oint_{S_3} \vec{E} \cdot \mathrm{d}\vec{S_3} \, \vec{u} 
= \frac{Q_\text{int}}{\varepsilon_0}
$$

avec $\mathrm{d}S_2 >= 0$.

---

On développe les intégrales et on obtient

$$
2 E \pi r^2 = \frac{Q_\text{int}}{\varepsilon_0}
$$

---

Ensuite on substitue Qint

$$
2 E \pi r^2 = \frac{\pi r^2 Q}{\varepsilon_0 S} \quad \text{avec } S \text{ la surface totale de la plaque.}
$$

---

On simplifie le tout pour isoler E

$$
E = \frac{Q}{2 \varepsilon_0 S}
$$

---

On obtient le résultat final suivant :

$$
\vec{E} = \frac{Q}{2 \varepsilon_0 S} \vec{e_y}
$$








### Trouvons l’angle incident final :

On considère le faisceau de particules arrivant avec un angle initial de $\theta$ avec $\vec{e_y}$ et une vitesse initiale $\vec{v}_0$.

On considère que les particules sont soumises au champ créé par la plaque à partir du point $(x_0, y_0)$.

D'après les équations d'une particule de charge $q$ et de masse $m$ dans un champ électrique :

$$
\begin{cases}
x(t) = \ v_0 t\cdot\vec{e_x}\ + x_0 \quad \text{(1)} \\ 
y(t) = \frac{q}{2m} E t^2\cdot\vec{e_y}\ + v_0 t\cdot\vec{e_y}\ + y_0 \quad \text{(2)}
\end{cases}
$$

---

A présent, calculons l'équation de la trajectoire.

À partir de (1), on exprime $t$ :

$$
t = \frac{x - x_0}{v_0 \cdot \vec{e_x}}
$$


Substituons cette valeur de $t$ dans (2) pour obtenir $y(x)$ :

$$
y(x) = \frac{q}{2m} \left( \frac{x - x_0}{v_0\cdot\vec{e_x}} \right)^2 E\cdot\vec{e_y} 
+ v_0\cdot\vec{e_y} \left( \frac{x - x_0}{v_0\cdot\vec{e_x}} \right) + y_0
$$


Simplifions :

$$
y(x) = \frac{q}{2m} \left( \frac{x - x_0}{v_0 \sin \theta} \right)^2 E 
+ v_0 \cos \theta \left( \frac{x - x_0}{v_0 \sin \theta} \right) + y_0
$$

---

On dérive $y(x)$ et on obtient

$$
y'(x) = \frac{q E}{m (v_0 \sin \theta)^2} \left( x - x_0 \right) + \frac{1}{\tan \theta}
$$

---

A présent, on pose $x_s$ la solution de l’équation $y(x) = 0$.  
Alors l’équation de la tangente en $x_s$ dans le repère $(O_s, \vec{e_x}, \vec{e_y})$ où $O_s$ est le point $(x_s, 0)$, est :

$$
y = y'(x_s) x
$$


On en déduit : 

$$
\tan\left(\frac{\pi}{2} - \alpha\right) = y'(x_s)
$$

En utilisant la définition de la tangente :

$$
\frac{\sin\left(\frac{\pi}{2} - \alpha\right)}{\cos\left(\frac{\pi}{2} - \alpha\right)} = y'(x_s)
$$

D'après les identités trigonométriques :

$$
\frac{\cos\alpha}{\sin\alpha} = y'(x_s)
$$

En isolant $\alpha$ :

$$
\alpha = \tan^{-1}\left(\frac{1}{y'(x_s)}\right)
$$
