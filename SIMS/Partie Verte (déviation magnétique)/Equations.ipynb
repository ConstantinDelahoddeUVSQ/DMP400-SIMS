On veut étudier la trajectoire d'une particule dans un champ magnétique suivant Bz.
Pour se faire on commence par un PFD. On remarque que l'unique force a laquelle la particule est soumise est la force de Lorentz. On a alors : $$\sum \vec{F}_{\text{ext}} = m \vec{a} = q \vec{v} \times \vec{B}$$
En appliquant la formule du produit vectoriel, on obtient 
$$
\vec{F}_{\text{ext}} =
q \big[ (v_y B_z - v_z B_y) \vec{x} - (v_x B_z - v_z B_x) \vec{y} + (v_x B_y - v_y B_x) \vec{z} \big].
$$
Or les composantes du champ sont nulles selon x et y donc on obtient 
$$
q B_z \big( v_y \vec{x} - v_x \vec{y} \big)= m \big( a_x \vec{x} + a_y \vec{y} + a_z \vec{z} \big) 
$$
Lorsqu'on projette on obtient :
$$
\dot{V}_x = \frac{q B_z}{m} V_y
$$

$$
\dot{V}_y = -\frac{q B_z}{m} V_x
$$

$$
\dot{V}_z = 0
$$

On en déduit
$$
V_x = -\frac{m}{q B_z} \dot{V}_y
$$
Et en dérivant
$$
\dot{V}_x = -\frac{m}{q B_z} \ddot{V}_y
$$
Alors on remplace et on obtient 
$$
\frac{q B_z}{m} V_y = -\frac{m}{q B_z} \ddot{V}_y
$$
$$
\frac{m}{q B_z} \ddot{V}_y + \frac{q B_z}{m} V_y = 0
$$

Immédiatement on reconnait une équation différentielle de la forme:  y''+ ay = 0 dont on connait la réponse.
$$
\ddot{V}_y + \left(\frac{q B_z}{m}\right)^2 V_y = 0
$$

$$
V_y (t) = A \cos\left(\frac{q B_z}{m} t\right) + B \sin\left(\frac{q B_z}{m} t\right)
$$

Enfin on dérive l'expression de Vy(t) pour obtenir Vx(t)
$$
V_x (t) = A \sin\left(\frac{q B_z}{m} t\right) - B \cos\left(\frac{q B_z}{m} t\right)
$$


Or on fournit à notre particule une vitesse initiale selon y seulement donc on a que Vy(t=0)=V0 et Vx(t=0)=0. En appliquant ces conditions on peut facilement trouver les valeurs de A et B. En effet: 
$$
V_y(t=0) = A \cos\left(0\right) + B \sin\left(0\right) = A = V_0
$$

$$
V_x(t=0) = A \sin\left(0\right) - B \cos\left(0\right) = B = 0
$$

Enfin , on remplace A et B.
$$
V_y(t) = V_0 \cos\left(\frac{q B_z}{m} t\right)
$$

$$
V_x(t) = V_0 \sin\left(\frac{q B_z}{m} t\right)
$$

Puis on intègre, en considérant que la particule est lancée depuis l'origine du repère cartésien, donc x_0=0 et y_0=0. On obtient :
$$
x(t) = \frac{V_0 m}{q B_z} \big[1 - \cos\left(\frac{q B_z}{m} t\right)\big]
$$

$$
y(t) = \frac{V_0 m}{q B_z} \sin\left(\frac{q B_z}{m} t\right)
$$


On se sert de l'équation temporelle de x pour isoler t en fonction de x :
$$
t = \frac{m}{q B_z} \arccos\left(1 - \frac{q B_z}{V_0 m} x\right)
$$



Enfin on réinsère dans l'équation temporelle de y et on obtient l'équation de la trajectoire de la particule : 
$$
y(x) = \frac{V_0 m}{q B_z} \sin\left(\arccos\left(1 - \frac{q B_z}{V_0 m} x\right)\right)
$$
